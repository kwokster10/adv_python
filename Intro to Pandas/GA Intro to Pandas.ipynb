```python
# How many songs were released in 1981
df2[df2["Release Year"] == "1981"]["Song Clean"].count()
```

```python
# What is the earliest release year in the data?
import re
df2["Release Year"].unique()
```

```python
def regex_year(value):
    if np.isnan(value):
        re.match(r"(19|20)\d{2}", value)

df2["Release Year"].str().apply(regex_year)
```

```python
def clean_year(value):
    # takes a value from df2["Release Year"] and changes it into a valid int year
    try: 
        value = int(value)
    except ValueError:
        return None
    else: 
        if value >= 1950:
            return value
        else: 
            return None
```

```python
df2["Clean Year"] = df2["Release Year"].apply(clean_year)
```

```python
df2["Clean Year"].min()
```

```python
# Lamda Version
df2["Song Clean"].apply(lambda val: val.lower())
```

```python
df3 = pd.read_csv("https://raw.githubusercontent.com/suneel0101/lesson-plan/master/crunchbase_monthly_export.csv")
```

```python
df3
```

```python
# Rename the column names accordingly
df3.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
df3.columns
```

```python
df3["funding_total_usd"] = df3["funding_total_usd"].apply(lambda val: int(re.sub(r',', '', val.strip().replace('-','0'))))
```

```python
df3["funding_total_usd"].max()
```

```python
df3.describe()
```

```python
# Construct a pivot table that indexes on region and shows the total funding amounts by region
pd.pivot_table(df3, index=["region"], aggfunc=np.sum, values=["funding_total_usd"])
```

```python
# What is the average number of funding rounds for companies in NYC? How does that compare to SF?
pd.pivot_table(df3[df3["city"] == "New York"], index=["city"], aggfunc=np.average, values=["funding_rounds"])
```

```python
pd.pivot_table(df3[df3["city"] == "San Francisco"], index=["city"], aggfunc=np.average, values=["funding_rounds"])
```

```python
pd.pivot_table(df3.loc[(df3["city"] == "New York") | (df3["city"] == "San Francisco")], index=["region"], aggfunc=np.average, values=["funding_rounds"])
```

```python
# What are the top 3 markets with the highest average funding total per company?
pd.pivot_table(df3, index=["market"], aggfunc=np.average, values=["funding_total_usd"]).sort_values(["funding_total_usd"], ascending=False)[:3]
```

```python
# How many companies have Games as a category? What's their average funding total? Using a pivot table, 
# what's the average funding total of those companies that have Games as a category, broken down by Region?
df3["contains_games"] = df3["category_list"].str.contains("Games")
pd.pivot_table(df3[df3["contains_games"] == True], index=["city"], aggfunc=np.average, values=["funding_total_usd"])
```
